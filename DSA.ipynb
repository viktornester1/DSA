# **LINKED LISTS**

## Singly linked lists

```python
class SinglyNode:
  def __init__(self, val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)
```

```python
Head = SinglyNode(1)
A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(7)

Head.next = A
A.next = B
B.next = C

print(Head)
```

```python
# Traversal of list - O(n)
curr = Head

while curr:
  print(curr)
  curr = curr.next
```

```python
# Displaying limked list - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' -> '.join(elements))

display(Head)
```

```python
# Search for a node value - O(n)
def search(head, val):
  curr = head
  while curr:
    if val == curr.val:
      return True
    curr = curr.next

  return False

search(Head, 1)
```

## Doubly linked lists

```python
class DoublyNode:
  def __init__(self, val, next = None, prev = None):
    self.val = val
    self.next = next
    self.prev = prev

  def __str__(self):
    return str(self.val)
```

```python
head = tail = DoublyNode(1)
print(tail)
```

```python
# Display - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' <-> '.join(elements))

display(head)
```

```python
# Insert at the beginning - O(1)
def insert_at_the_beginning(head, tail, val):
  new_node = DoublyNode(val, next=head)
  head.prev = new_node
  return new_node, tail

head, tail = insert_at_the_beginning(head, tail, 3) # призначаєм елементи
display(head)
```

```python
# Insert at the end - O(1)
def insert_at_the_end(head, tail, val):
  new_node = DoublyNode(val, prev=tail)
  tail.next = new_node
  return head, new_node

head, tail = insert_at_the_end(head, tail, 7)
display(head)
```

# **Binary Search**

```python
A = [-3, -1, 0, 1, 4, 7]

# Naive O(n) searching
if -1 in A:
  print(True)
```

```python
# Traditional Binary search - looking up if value is in array
# Time O(log n)
# space O(1)
def binary_search(arr, target):

  N = len(arr)
  L = 0
  R = N-1

  while L <= R:  # even if equal still need a final check if thats what we look for
    M = L + ((R-L) // 2) # M = (L+R) // 2

    if arr[M] == target:
      return True
    elif target < arr[M]:
      R = M - 1 # search in left side
    else:
      L = M + 1 # search in right side
  return False


binary_search(A, 0)
```

```python
# Based on the condition
B = [False, False, False, False, True, True, True, True]
def binary_search_condition(arr):
  N = len(arr)
  L = 0
  R = N - 1
  while L < R:
    M = (L+R) // 2

    if B[M]:
      R = M
    else:
      L = M + 1

  return L

binary_search_condition(B) # Searches for beginning of True
```

# **Binary Trees**

```python
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
```

```python
#        1
#     2      3
#   4   5  10
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
```

```python
# Recursive Pre Order Traversal (DFS) Time: O(n), Space: O(n)
def pre_order(node):
  if not node:
    return

  print(node)
  pre_order(node.left)
  pre_order(node.right)

pre_order(A)
```

```python
# Recursive In Order Traversal (DFS) Time: O(n), Space: O(n)
def in_order(node):
  if not node:
    return

  pre_order(node.left)
  print(node)
  pre_order(node.right)

pre_order(A)
```

```python
# Recursive Post Order Traversal (DFS) Time: O(n), Space: O(n)
def in_order(node):
  if not node:
    return

  pre_order(node.left)
  pre_order(node.right)
  print(node)

pre_order(A)
```

```python
# Iterative Pre Order Traversal (DFS) Time: O(n), Space: O(n)
def pre_order_iterative(node):
  stk = [node] # the root

  while stk:
    node = stk.pop()
    print(node)
    if node.right: stk.append(node.right)
    if node.left: stk.append(node.left) # left will be on top, will force us on left side

pre_order_iterative(A)
```

```python
# Level Order Traversal (BFS) Time: O(n), Space O(n)
from collections import deque

def level_order(node):
  q = deque()
  q.append(node)

  while q:
    node = q.popleft()
    print(node)
    if node.left: q.append(node.left)
    if node.right: q.append(node.right)


level_order(A)
```

```python
# Check If Value Exists (DFS) Time: O(n), Space: O(n)
def search(node, target):
  if not node:
    return False

  if node.val == target:
    return True

  return search(node.left, target) or search(node.right, target)

search(A, 5)
```

```python
# Time: O(log n), Space: O(log n)
def search_bst(node, target):
  if not node:
    return False

  if node.val == target:
    return True

  if target < node.val: return search_bst(node.left, target)
  else: return search_bst(node.right, target)

search_bst(A2, 8)
```

# **Heaps & Priority Queues**

```python
# Build Min Heap (Heapify)
# Time: O(n), Space: O(1)

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]

import heapq
heapq.heapify(A)

A
```

```python
# Heap Push (Insert element)
# Time: O(log n)

heapq.heappush(A, 4)

A
```

```python
# Heap Pop (Extract min)
# Time: O(log n)

minn = heapq.heappop(A)

A, minn
```

```python
# Heap Sort
# Time: O(n log n)
# Space: O(n) -> O(1) is possible but harder to do
def heapsort(arr):
  heapq.heapify(arr)
  n = len(arr)
  new_list = [0] * n

  for i in range(n):
    minn = heapq.heappop(arr)
    new_list[i] = minn

  return new_list


heapsort([1, 3, 5, 7, 9, 2, 4, 6, 8, 0])
```

```python
# Heap Push Pop: Time: O(log n)
heapq.heappushpop(A, 99)
A
```

```python
# Peak at Min: Time O(1)
A[0]
```

```python
# Max Heap

B = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
n = len(B)

for i in range(n):
  B[i] = -B[i]

heapq.heapify(B)

B
```

```python
largest = -heapq.heappop(B) # heapq has no max heap

largest
```

```python
heapq.heappush(B, -7) # insert 7
B
```

```python
# Build heap from scratch - Time: O(n log n) slower than heapify

C = [-5, 4, 2, 1, 7, 0, 3]

heap = []

for x in C:
  heapq.heappush(heap, x)
  #print(heap)
  print(heap, len(heap)) # Check size of heap
```

```python
# Putting tuples of items on the heap

D = [5, 4, 3, 5, 4, 3, 5, 5, 4]

from collections import Counter

counter = Counter(D)
counter # 4 fives, 3 fours, 2 threes
```

```python
heap = []

for k, v in counter.items():
  heapq.heappush(heap, (v, k)) # sorts by v first, sorts by k if tie

heap
```

```python
# Convert Array of Edges -> Adjacency List
from collections import defaultdict

D = defaultdict(list)

for u, v in A:
  D[u].append(v) # D[3].append(6)
  # If graph is undirected
  D[v].append(u)

D
```

```python
# DFS with Recursion - O(V + E) where V is the number of nodes and E is the number of edges

def dfs_recursive(node):
  print(node)
  for nei_node in D[node]:
    if nei_node not in seen:
      seen.add(nei_node)
      dfs_recursive(nei_node)

source = 0
seen = set()
seen.add(source)
dfs_recursive(source)
```

```python
# Iterative DFS with Stack O(V + E)

source = 0
seen = set()
seen.add(source)
stack = [source]
while stack:
  node = stack.pop()
  print(node)
  for nei_node in D[node]:
    if nei_node not in seen:
      seen.add(nei_node)
      stack.append(nei_node)
```

```python
# BFS (Queue) - O(V + E)

source = 0

from collections import deque

seen = set()
seen.add(source)
q = deque()
q.append(source)

while q:
  node = q.popleft()
  print(node)
  for nei_node in D[node]:
    if nei_node not in seen:
      seen.add(nei_node)
      q.append(nei_node)
```
